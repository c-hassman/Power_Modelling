```python
import pypsa, numpy as np, pandas as pd
```

```python
# Set up grid. In future, more complex model iterations, these things can be loaded by CSV
# Additionally, there are many other attributes of generators which should be considered
# There is a logging example in the pypsa documentation. Should definietly go through
# Also a lot of good visualization tools: would be good if I could get a way to automatically 
# run scenarios and produce a report 

# Marginal Costs in USD/MWH
marginal_costs = {"Solar": 3,
                  "Wind" : 4,
                  "Nuclear" : 5,
                  "SE_Gas" : 12, 
                  "Base_Gas" : 21,
                  "Peak_1" : 35,
                  "Peak_2" : 40,
                  "Peak_3" : 50, 
                  "Oil" : 70}

# Power Plant Capacities, nominal in MW
power_plant_p_nom = {"West" : {"Solar": 0,
                               "Wind" : 25000,
                               "Nuclear" : 0,
                               "SE_Gas" : 0, 
                               "Base_Gas" : 10000,
                               "Peak_1" : 5000,
                               "Peak_2" : 0,
                               "Peak_3" : 5000, 
                               "Oil" : 0}, 
                     
                     "North" : {"Solar": 0,
                               "Wind" : 0,
                               "Nuclear" : 0,
                               "SE_Gas" : 0, 
                               "Base_Gas" : 10000,
                               "Peak_1" : 5000,
                               "Peak_2" : 5000,
                               "Peak_3" : 5000, 
                               "Oil" : 0}, 
                     
                     "South" : {"Solar": 7000,
                               "Wind" : 5000,
                               "Nuclear" : 0,
                               "SE_Gas" : 0, 
                               "Base_Gas" : 10000,
                               "Peak_1" : 5000,
                               "Peak_2" : 5000,
                               "Peak_3" : 5000, 
                               "Oil" : 0}, 
                     
                     "Houston" : {"Solar": 0,
                               "Wind" : 0,
                               "Nuclear" : 5000,
                               "SE_Gas" : 5000, 
                               "Base_Gas" : 10000,
                               "Peak_1" : 5000,
                               "Peak_2" : 5000,
                               "Peak_3" : 5000, 
                               "Oil" : 5000}
                    }
                     
# Tranmission Capacities <- don't have great estimates for this yet

# commented out duplicate lines, because i will make lines bidirectional later

transmission = {"West" : {"North" : 5000,
                         "South" : 5000},
               "North" : {# "West" : 0, # Not sure about this
                          "Houston" : 5000},
               "South" : {# "West" : 0,
                          "North" : 3500,
                          "Houston" : 3500},
               #"Houston" : {"North" : 1500,
               #             "South" : 1500}                          
              }

# Loads 
loads = {"West" : 5000,
         "North" : 30000,
         "South" : 15000, 
         "Houston" : 15000}
```

# Single Bidding Zone with Fixed Load, One period
Only consider Houston
Inelastic load

```python
# Single Bidding Zone with Fixed Load, One period
# Only consider Houston
# Inelastic load

zone = "Houston"
network = pypsa.Network()
network.add("Bus", zone) # Add the busses

# Add Generation with Marginal Costs
# In future, more complex iterations, different marginal costs can be loaded for zone, 
# up to individual power plants. Might be good to do a micro-grid model
for tech in power_plant_p_nom[zone]: 
    # power_plant_p_nom[zone] returns all of the nomimal capacities
    # for the differenct techs in that zone
    network.add("Generator", 
                "{} {}".format(zone, tech),
                bus = zone, # assign the generator to that zone
                p_nom = power_plant_p_nom[zone][tech], # Assign marginal nominal capacity
                marginal_cost = marginal_costs[tech]) # assign marginal costs
    
network.add("Load", # Add load for the zone
           "{} load".format(zone),
           bus = zone, 
           p_set = loads[zone]) # assign load value
```

```python
# run optimization to determine market dispatch
network.lopf()
```

```python
network.loads_t.p # print the load active power consumption 
```

```python
# print the gerator active power (P) dispatch
network.generators_t.p
```

```python
network.buses_t.marginal_price # print clearing price
```

# Four connected bidding zones, connected by transmission, one period

From: https://pypsa.org/examples/simple-electricity-market-examples.html

"In this example we have bidirectional transmission capacity between two bidding zones. The power transfer is treated as controllable (like an A/NTC (Available/Net Transfer Capacity) or HVDC line). Note that in the physical grid, power flows passively according to the network impedances."

So HVDC lines can be bidirectional? Even in the US?


```python
network = pypsa.Network()
zones = ['West', 'South', 'North', 'Houston']

for zone in zones:
    network.add("Bus", zone)
    
    for tech in power_plant_p_nom[zone]:
        network.add("Generator",
                    "{} {}".format(zone, tech),
                    bus = zone,
                    p_nom = power_plant_p_nom[zone][tech],
                    marginal_cost = marginal_costs[tech])
    
    print(zone, loads[zone])
    network.add("Load", 
                "{} load".format(zone),
                bus = zone, 
                p_set = loads[zone])
    
    # add tranmission as controllable Link <- not sure about this
    if zone not in transmission: 
        continue
    
    for other_zone in zones:
        #print(zone, other_zone)
        
        if other_zone not in transmission[zone]:
            continue
        # Link is by default unidirectional, so have to set p_min_pu = -1
        # this allows bidrections (i.e. negative) flow
        network.add("Link", 
                    "{} - {} link".format(zone, other_zone),
                   bus0 = zone, 
                   bus1 = other_zone, 
                   p_nom = transmission[zone][other_zone],
                   p_min_pu = -1)
```

```python
# Run Optimizer
network.lopf()
```

```python
# Print loads
# Must run the optimizer first
network.loads_t.p
```

```python
#print(type(network.generators_t.p))

l = network.generators_t.p.values # get generator values as an array


techs = ["Solar", "Wind", "Nuclear", "SE_Gas", "Base_Gas", "Peak_1", "Peak_2", "Peak_3", "Oil"]
zones = ['West', "South", "North", "Houston"]

gen = pd.DataFrame(l.reshape(len(zones), 9), columns = techs) # convert array to matrix then to dataframe.
        # 9 is the number of techs
gen.index = zones # Set zones as index
gen
```

```python
# Print linking tranmission flows
network.links_t.p0
```

```python
# print clearing price
network.buses_t.marginal_price
```

```python
# Print shadow prices
network.links_t.mu_lower
```

```python
# to add a price sensitive load, we can use a dummy negative-dispatch generator

# Datacenter case study has a ineslastic base load but also low marginal utility 
# variable load: extra computations when its cheap to compute
network.add("Generator", 
            "West Texas Datacenter load",
            bus = "West", 
            p_max_pu = 0, # max power output as % of nominal
            p_min_pu = -1, # max power draw as % of nominal
            p_nom = 8000,
            marginal_cost = 10) # really marginal utility

network.add("Generator",
            "Houston Industrial load", 
             bus ="Houston", 
             p_max_pu = 0, 
             p_min_pu = -1, 
             p_nom = 5000, 
             marginal_cost = 25)
```

# Multiple Time Periods

Here we are finally in it. In this model I will change load based on a discount to peak, as well as change the renewable generation as a percentage of capacity


```python
network = pypsa.Network()
zones = ['West', 'South', 'North', 'Houston']

# snapshots are zero-indexed
network.set_snapshots(range(10))

# Add zones to network
for zone in zones:
    # ADD BUSSES
    network.add("Bus", zone)
    
    # ADD GENERATION
    for tech in power_plant_p_nom[zone]:
        network.add("Generator", 
                    "{} {}".format(zone, tech),
                    bus = zone, 
                    p_nom = power_plant_p_nom[zone][tech], 
                    marginal_cost = marginal_costs[tech], 
                    # variability of renewable gen
                    # can't figure out how to do multiple
                    p_max_pu = ([1,0.9,0.75,0.5,0.3,0.3,0.5,0.75,0.9,1] if tech == "Wind" else 1),
                                #(if tech == "Wind":
                                #   [1,0.9,0.75,0.5,0.3,0.3,0.5,0.75,0.9,1] 
                                #elif tech == "Solar":
                                #   [0.1,0.2,0.5,0.75,1,1,1,0.7,0.5,0.2] 
                                #else 1)                            
                   )
        
    # ADD LOAD
    network.add("Load",
                "{} load".format(zone),
                bus = zone,
                #p_set = loads[zone] + np.array([-3000, -3000, -2000, -1500, -500, 0,
                #                                -500, -1500, -2000, -3000])
                p_set = loads[zone] * np.array([0.5, 0.6, 0.7, 0.8, 0.9, 1, 1, 0.9, 0.8, 0.7])
                   
               )
        
    # ADD TRANSMISSION
    if zone not in transmission:
        continue
            
    for other_zone in zones:
        if other_zone not in transmission[zone]:
            continue
        
            # again we set tranmission to be bidirectional
        network.add("Link", 
                        "{} - {}".format(zone, other_zone),
                        bus0 = zone, 
                        bus1 = other_zone, 
                        p_nom = transmission[zone][other_zone],
                        p_min_pu = -1)
        
```

```python
network.lopf()
```
